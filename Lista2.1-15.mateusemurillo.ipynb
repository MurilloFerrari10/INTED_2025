```python
5
salario_atual = float(input("Digite o salário atual: "))

if salario_atual <= 2000:
    percentual_aumento = 0.10  # 10%
elif salario_atual <= 5000:
    percentual_aumento = 0.07  # 7%
else:
    percentual_aumento = 0.05  # 5%

novo_salario = salario_atual * (1 + percentual_aumento)

print(f"Salário atual: R${salario_atual:.2f}")
print(f"Percentual de aumento: {percentual_aumento * 100:.0f}%")
print(f"Novo salário: R${novo_salario:.2f}")
```

```python
11
soma = 0

while True:
    numero = float(input("Digite um número para somar (ou 0 para encerrar): "))
    if numero == 0:
        break
    soma += numero

print(f"A soma total dos números é: {soma}")
```

```python
13
soma = 0
numero = 1

print("Somando números sequenciais até a soma ultrapassar 100:")

while soma <= 100:
    print(f"Adicionando: {numero}")
    soma += numero
    numero += 1

print(f"A soma final é: {soma}")
```

```python
14
soma_idades = 0
quantidade_idades = 0

print("Digite as idades (digite um número negativo para encerrar):")

while True:
    idade = int(input("Idade: "))

    if idade < 0:
        break

    soma_idades += idade
    quantidade_idades += 1

if quantidade_idades > 0:
    media = soma_idades / quantidade_idades
    print(f"A média das idades digitadas é: {media:.2f}")
else:
    print("Nenhuma idade válida foi digitada para calcular a média.")
```

```python
15
numero_inicial = int(input("Digite o número inicial para a contagem regressiva: "))

print(f"Iniciando contagem regressiva a partir de {numero_inicial}...")

while numero_inicial >= 0:
    print(numero_inicial)
    numero_inicial -= 1

print("Contagem regressiva finalizada!")
```
