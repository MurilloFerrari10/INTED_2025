```python
19.
maiores_de_18 = 0

for i in range(5):
  try:
    idade = int(input(f"Digite a {i+1}ª idade: "))
    if idade > 18:
      maiores_de_18 += 1
  except ValueError:
    print("Idade inválida. Por favor, digite um número inteiro.")

print(f"Das 5 idades digitadas, {maiores_de_18} são maiores de 18 anos.")
```

```python
20.
maior_numero = None

for i in range(1, 8):
  try:
    numero = int(input(f"Digite o {i}º número: "))
    if maior_numero is None or numero > maior_numero:
      maior_numero = numero
  except ValueError:
    print("Entrada inválida. Por favor, digite um número inteiro.")
    continue

if maior_numero is not None:
  print(f"O maior número digitado é: {maior_numero}")
else:
  print("Nenhum número válido foi digitado.")
```

```python
21.
tarefa_mais_longa = ""
maior_tamanho = 0

for i in range(1, 5):
  nome_tarefa = input(f"Digite o nome da {i}ª tarefa: ")
  if len(nome_tarefa) > maior_tamanho:
    maior_tamanho = len(nome_tarefa)
    tarefa_mais_longa = nome_tarefa

print(f"A tarefa com o nome mais longo é: '{tarefa_mais_longa}' (com {maior_tamanho} caracteres).")
```

```python
23.
positivos = 0
negativos = 0
zeros = 0

for i in range(1, 11):
  try:
    numero = int(input(f"Digite o {i}º número: "))
    if numero > 0:
      positivos += 1
    elif numero < 0:
      negativos += 1
    else:
      zeros += 1
  except ValueError:
    print("Entrada inválida. Por favor, digite um número inteiro.")
    continue

print(f"Números positivos: {positivos}")
print(f"Números negativos: {negativos}")
print(f"Números iguais a zero: {zeros}")
```

```python
24.
print("Digite 6 nomes para filtrar (que começam com 'A' ou 'C'):")

for i in range(1, 7):
  nome = input(f"Digite o {i}º nome: ")
  # Convert the name to uppercase for case-insensitive comparison
  nome_maiusculo = nome.upper()
  if nome_maiusculo.startswith('A') or nome_maiusculo.startswith('C'):
    print(f"Nome que começa com 'A' ou 'C': {nome}")
```

```python
27.
senha_correta = "python123"
max_tentativas = 3
tentativas = 0

print("Sistema de Login")

while tentativas < max_tentativas:
  senha_digitada = input(f"Digite a senha (tentativa {tentativas + 1}/{max_tentativas}): ")

  if senha_digitada == senha_correta:
    print("Login bem-sucedido! Bem-vindo(a).")
    break
  else:
    print("Senha incorreta.")
    tentativas += 1

if tentativas == max_tentativas:
  print("Você excedeu o número máximo de tentativas. Tente novamente mais tarde.")
```

```python
28.
soma_positivos = 0

print("Digite 10 números:")

for i in range(1, 11):
  try:
    numero = int(input(f"Digite o {i}º número: "))
    if numero < 0:
      print("Número negativo ignorado.")
      continue  # Ignora números negativos

    soma_positivos += numero
  except ValueError:
    print("Entrada inválida. Por favor, digite um número inteiro.")
    continue # Ignora entradas inválidas

print(f"A soma dos números positivos digitados é: {soma_positivos}")
```

```python
29.
soma = 0
contador = 0

print("Acumulador: Digite números para somar (o laço encerra quando a soma ultrapassar 200).")

while True:
  try:
    numero = int(input(f"Digite o {contador + 1}º número: "))
    soma += numero
    contador += 1
    print(f"Soma atual: {soma}")

    if soma > 200:
      print(f"A soma ({soma}) ultrapassou 200. Encerrando o acumulador.")
      break
  except ValueError:
    print("Entrada inválida. Por favor, digite um número inteiro.")
    continue

print(f"Total final acumulado: {soma}")
print(f"Quantidade de números digitados: {contador}")
```

```python
30.
contador_palavras = 0

print("Programa de contagem de palavras (digite 'fim' para encerrar).")

while True:
  palavra = input("Digite uma palavra: ").lower() # Converte para minúsculas para padronizar

  if palavra == "fim":
    print("\n'fim' digitado. Encerrando a contagem.")
    break
  else:
    contador_palavras += 1
    print(f"Palavra '{palavra}' registrada. Total atual: {contador_palavras}")

print(f"Total de palavras válidas informadas: {contador_palavras}")
```
